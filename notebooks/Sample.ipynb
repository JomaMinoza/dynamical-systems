```python
import numpy as np
from math import sin, cos, pi
from plotdf import plotdf
```

```python
def f(x,r=1):
    return np.array([x[1],r-x[0]**2])

plotdf(f, # Function giving the rhs of the diff. eq. system
         np.array([-1,1]), # [xmin,xmax]
         np.array([-1,1]),# [ymin,ymax]
         [(1,0),(0,0)], # list of initial values for trajectories (optional)
         parameters={"r": 1}
       
      )
```

```python
def f(x,r=1):
    return np.array([x[1],1-x[0]**14])

plotdf(f, # Function giving the rhs of the diff. eq. system
         np.array([-2,2]), # [xmin,xmax]
         np.array([-2,2]),# [ymin,ymax]
         [(1,0),(.5,0)], # list of initial values for trajectories (optional)
         parameters={"r": 1}
       
      )
```

```python
def f(x,r=1):
    return np.array([x[1],1-2*cos(x[0])])

plotdf(f, # Function giving the rhs of the diff. eq. system
         np.array([-4,4]), # [xmin,xmax]
         np.array([-4,4]),# [ymin,ymax]
         [(1,0),(.5,0)], # list of initial values for trajectories (optional)
         parameters={"r": 1}
       
      )
```

```python
import matplotlib.pyplot as plt

x = np.linspace(-3,3,100)
y = 1 - 2*np.sin(x)

# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the function
plt.plot(x,y, 'g')

# show the plot
plt.show()
```
